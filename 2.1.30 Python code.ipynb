```python
import numpy as np
import matplotlib.pyplot as plt

u = np.array([1, 0])
A = np.array([[0.8, 0.3], [0.2, 0.7]])
x = [u]
k = np.arange(8)

while len(x) <= 7:
    u = np.dot(A, u)
    x.append(u)

x = np.array(x)

plt.figure(figsize=(6,4))
plt.plot(k, x[:, 0],label='u[: 0]')
plt.plot(k, x[:, 1],label='u[: 1]')


v = np.array([0,1])
A = np.array([[0.8, 0.3], [0.2, 0.7]])
x = [v]
k = np.arange(8)

for j in range(7):
    v = np.dot(A, v)
    x.append(v)

x = np.array(x)

plt.plot(k, x[:, 0],ls = '--',color='tab:blue',label='v[: 0]')
plt.plot(k, x[:, 1],ls = '--',color='tab:orange',label='v[: 1]')
plt.legend()
plt.show()
```
